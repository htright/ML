# 자료불러오기

```python
import pandas as pd
```

```python
df = pd.read_excel('아파트(매매)_실거래가_2018.xlsx')
df = df[15:]
col = list(df.iloc[0])
df.columns = col
df = df[1:]
df
```

```python
for i in range(2018, 2022):
    df_plus = pd.read_excel(f'아파트(매매)_실거래가_{i}.xlsx',thousands=',')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df = df.append(df_plus)
    
df
```

```python
df.rename(columns = {'거래금액(만원)':'거래금액'}, inplace = True)
```

```python
df.tail()
```

```python
df['거래금액2'] = df.거래금액.str.replace(',', '').astype('int64')
```

```python
df.columns
```

```python
df.describe()
```

```python
df
```

```python
cols = ['시군구', '번지']
df['주소'] =df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df
```

```python
df3 = df.groupby(['주소', '계약년월'])[['거래금액2']].mean()
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3 = df3.reset_index()
df3
```

```python
df3.to_excel('거래금액v1.xlsx')
```

## 201801 거래내역 있는 주소만 추출

```python
df3.dtypes
```

```python
df3['계약년월'] = df3.계약년월.astype('int64')
```

```python
df4 = df3[df3['계약년월'] == 201801]
df4
```

```python
df4.rename(columns = {'거래금액2':'기준'}, inplace = True)
```

```python
print(df4)
print(df4.dtypes)
```

## 2018년 1월 거래가격 vlookup

```python
df5= pd.merge(df3[['주소', '계약년월', '거래금액2']],df4[['주소', '기준']], on='주소' ,how='left')
```

```python
df5
```

```python
df5.isnull().sum()
```

## 2018년 거래내역 없는 주소 제거

```python
df5 = df5.dropna(axis=0)
```

```python
df5.isnull().sum()
```

## 가격지수

```python
df5['가격지수'] = df5['거래금액2'] / df5['기준'] * 100
```

```python
df5
```

```python
df6 = df5[['주소', '계약년월', '가격지수']].reset_index()
```

```python
df6
```

```python
df6.to_excel("가격지수.xlsx", index = False)
```
