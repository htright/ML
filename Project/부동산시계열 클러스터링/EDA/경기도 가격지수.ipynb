```python
import pandas as pd
```

```python
df = pd.read_excel('아파트(매매)_실거래가_2018.xlsx')
df = df[15:]
col = list(df.iloc[0])
df.columns = col
df = df[1:]
df
```

```python
for i in range(2018, 2022):
    df_plus = pd.read_excel(f'아파트(매매)_실거래가_{i}.xlsx',thousands=',')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df = df.append(df_plus)
    
df
```

```python
df.rename(columns = {'거래금액(만원)':'거래금액'}, inplace = True)
```

```python
df.tail()
```

```python
df['거래금액2'] = df.거래금액.str.replace(',', '').astype('int64')
```

```python
cols = ['시군구', '번지']
df['주소'] =df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df
```

```python
df3 = df.groupby(['주소', '계약년월'])[['거래금액2']].mean()
```
